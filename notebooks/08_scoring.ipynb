```python
# 08 – Player Scoring & Ranking

Purpose:
- Score players within each role pool
- Rank candidates transparently using weighted logic

Inputs:
- Role-based CSVs (batting, bowling, all-rounders)

Outputs:
- Ranked candidate tables per role
```

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from config import *
import pandas as pd
```

```python
openers = pd.read_csv(PROCESSED_DIR / "openers.csv")
middle = pd.read_csv(PROCESSED_DIR / "middle_order.csv")
finishers = pd.read_csv(PROCESSED_DIR / "finishers.csv")
all_rounders = pd.read_csv(PROCESSED_DIR / "all_rounders.csv")

first_bowlers = pd.read_csv(PROCESSED_DIR / "first_bowlers.csv")
middle_bowlers = pd.read_csv(PROCESSED_DIR / "middle_bowlers.csv")
death_bowlers = pd.read_csv(PROCESSED_DIR / "death_bowlers.csv")
```

```python
openers["score"] = (
    openers["matches_played"] * 0.4 +
    openers["Powerplay"] * 0.4 +
    openers["strike_rate"] * 0.3 +
    openers["average"] * 0.2
)

top_openers = openers.sort_values("score", ascending=False).head(4)
top_openers[["batsman", "score"]]
```

```python
middle["score"] = (
    middle["matches_played"] * 0.4 +
    middle["average"] * 0.6 +
    middle["Middle"] * 0.3 +
    (100 - middle["dot_ball_pct"]) * 0.2
)

top_middle = middle.sort_values("score", ascending=False).head(4)
top_middle[["batsman", "score"]]
```

```python
finishers["score"] = (
    finishers["matches_played"] * 0.4 +
    finishers["Death"] * 0.5 +
    finishers["strike_rate"] * 0.3 +
    finishers["avg_death_balls_per_match"] * 0.3
)

top_finishers = finishers.sort_values("score", ascending=False).head(4)
top_finishers[["batsman", "score"]]
```

```python
all_rounders["score"] = (
    all_rounders["matches_played"] * 0.2 +
    all_rounders["batting_strike_rate"] * 0.3 +
    all_rounders["batting_average"] * 0.3 +
    (10 - all_rounders["economy"]) * 0.3 +
    all_rounders["wickets"] * 0.3
)

top_all_rounders = all_rounders.sort_values("score", ascending=False).head(4)
top_all_rounders[["player", "score"]]
```

```python
first_bowlers["score"] = (
    first_bowlers["powerplay_balls_bowled"] * 0.4 +
    (10 - first_bowlers["powerplay_economy"]) * 0.4 +
    first_bowlers["wickets"] * 0.4
)

top_pp_bowler = first_bowlers.sort_values("score", ascending=False).head(2)
```

```python
middle_bowlers["score"] = (
    middle_bowlers["middle_overs_balls_bowled"] * 0.4 +
    (10 - middle_bowlers["middle_overs_economy"]) * 0.4 +
    middle_bowlers["wickets"] * 0.3
)

top_mid_bowler = middle_bowlers.sort_values("score", ascending=False).head(2)
```

```python
death_bowlers["score"] = (
    death_bowlers["death_overs_balls_bowled"] * 0.4 +
    (12 - death_bowlers["death_overs_economy"]) * 0.4 +
    death_bowlers["wickets"] * 0.3
)

top_death_bowler = death_bowlers.sort_values("score", ascending=False).head(2)
```

```python
top_openers.to_csv(OUTPUTS_DIR / "tables/top_openers.csv", index=False)
top_middle.to_csv(OUTPUTS_DIR / "tables/top_middle.csv", index=False)
top_finishers.to_csv(OUTPUTS_DIR / "tables/top_finishers.csv", index=False)
top_all_rounders.to_csv(OUTPUTS_DIR / "tables/top_all_rounders.csv", index=False)
```
