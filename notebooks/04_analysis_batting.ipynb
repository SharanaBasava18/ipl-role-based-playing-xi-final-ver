```python
# 04 – Batting Role Analysis

Purpose:
- Identify batting roles using engineered features
- Classify players as Openers, Middle Order, Finishers

Input:
- data/processed/batting_features_final_v2.csv

Outputs:
- data/processed/openers.csv
- data/processed/middle_order.csv
- data/processed/finishers.csv
```

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from config import *
import pandas as pd
```

```python
batting = pd.read_csv(
    PROCESSED_DIR / "batting_features_final_v2.csv"
)

print(batting.shape)
batting.head()
```

```python
openers = batting[
    (batting["matches_played"] >= 50) &
    (batting["balls"] >= 300) &
    (batting["strike_rate"] >= 135) &
    (batting["average"] >= 28) &
    (batting["Powerplay"] >= 140) &
    (batting["avg_pp_balls_per_match"] >= 6)
].copy()

openers = openers.sort_values(
    "matches_played", ascending=False
)

openers.head()
```

```python
middle_order = batting[
    (batting["matches_played"] >= 80) &
    (batting["balls"] >= 600) &
    (batting["strike_rate"] >= 120) &
    (batting["average"] >= 30) &
    (batting["Middle"] >= 120) &
    (batting["avg_middle_balls_per_match"] >= 6) &
    (batting["avg_middle_balls_per_match"] >
     batting["avg_pp_balls_per_match"])
].copy()

middle_order = middle_order.sort_values(
    "matches_played", ascending=False
)

middle_order.head()
```

```python
finishers = batting[
    (batting["matches_played"] >= 80) &
    (batting["balls"] >= 400) &
    (batting["strike_rate"] >= 140) &
    (batting["Death"] >= 150) &
    (batting["avg_death_balls_per_match"] >= 3)
].copy()

finishers = finishers.sort_values(
    "matches_played", ascending=False
)

finishers.head()
```

```python
openers.to_csv(
    PROCESSED_DIR / "openers.csv",
    index=False
)
```

```python
middle_order.to_csv(
    PROCESSED_DIR / "middle_order.csv",
    index=False
)
```

```python
finishers.to_csv(
    PROCESSED_DIR / "finishers.csv",
    index=False
)
```
