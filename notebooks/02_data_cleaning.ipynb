```python
# 02 – Data Cleaning & Standardization

Purpose:
- Standardize delivery-level data
- Identify legal deliveries
- Assign match phases

Input:
- data/processed/deliveries_raw.csv

Output:
- data/processed/deliveries_cleaned.csv
```

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from config import *
import pandas as pd
```

```python
df = pd.read_csv(PROCESSED_DIR / "deliveries_raw.csv")

print("Raw shape:", df.shape)
df.head()
```

```python
def assign_phase(over):
    if over <= POWERPLAY_END_OVER:
        return "Powerplay"
    elif over <= MIDDLE_END_OVER:
        return "Middle"
    return "Death"

df["phase"] = df["over"].apply(assign_phase)

df["phase"].value_counts()
```

```python
df.to_csv(
    PROCESSED_DIR / "deliveries_cleaned.csv",
    index=False
)

print("deliveries_cleaned.csv created successfully")
```
