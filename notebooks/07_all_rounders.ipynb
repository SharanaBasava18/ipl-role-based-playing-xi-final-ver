```python
# 07 – All-Rounder Identification

Purpose:
- Identify players who contribute with both bat and ball
- Merge final batting and bowling datasets

Inputs:
- data/processed/batting_features_final_v2.csv
- data/processed/bowling_metrics.csv

Output:
- data/processed/all_rounders.csv
```

```python
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().parent))

from config import *
import pandas as pd
```

```python
batting = pd.read_csv(
    PROCESSED_DIR / "batting_features_final_v2.csv"
)

bowling = pd.read_csv(
    PROCESSED_DIR / "bowling_metrics.csv"
)

print("Batting:", batting.shape)
print("Bowling:", bowling.shape)
```

```python
batting = batting.rename(columns={"batsman": "player"})
bowling = bowling.rename(columns={"bowler": "player"})
```

```python
all_rounders = pd.merge(
    batting,
    bowling,
    on="player",
    how="inner"
)

print("All-rounders:", all_rounders.shape)
all_rounders.head()
```

```python
all_rounders = all_rounders.rename(columns={
    # batting side
    "balls_x": "batting_balls",
    "runs": "batting_runs",
    "strike_rate": "batting_strike_rate",
    "average": "batting_average",

    # bowling side
    "balls_y": "bowling_balls"
})
```

```python
all_rounders = all_rounders[
    (all_rounders["batting_balls"] >= 200) &        # batting involvement
    (all_rounders["bowling_balls"] >= 200)          # bowling involvement
].copy()
```

```python
all_rounders.to_csv(
    PROCESSED_DIR / "all_rounders.csv",
    index=False
)

print("all_rounders.csv created successfully")
```
