```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

BASE_DIR = Path("..").resolve().parent
PROCESSED_DIR = BASE_DIR / "data" / "processed"

plt.style.use("seaborn-v0_8-darkgrid")
```

```python
playing_xi = pd.read_csv(PROCESSED_DIR / "final_playing_xi.csv")
playing_xi
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(
    data=playing_xi,
    x="role",
    order=playing_xi["role"].value_counts().index
)

plt.title("Role Distribution in Final Playing XI")
plt.xlabel("Role")
plt.ylabel("Number of Players")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
foreign_players = {
    "JC Buttler", "AB de Villiers", "KA Pollard",
    "SP Narine", "DJ Bravo", "AD Russell",
    "JM Bairstow", "DA Warner", "RG Maxwell",
    "F du Plessis", "N Pooran", "Rashid Khan",
    "SL Malinga"
}

playing_xi["Nationality"] = playing_xi["player"].apply(
    lambda x: "Overseas" if x in foreign_players else "Indian"
)

plt.figure(figsize=(6, 4))
sns.countplot(
    data=playing_xi,
    x="Nationality"
)

plt.title("Indian vs Overseas Players in Final XI")
plt.xlabel("")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
